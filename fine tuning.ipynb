```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("C:\\Users\\ADMIN\\Desktop\\New folder\\flights.csv")
df.head()
```

```python
df['AvgTicketPrice'] = df['AvgTicketPrice'].replace('[\$,]', '', regex=True).astype(float)
df['DistanceKilometers'] = df['DistanceKilometers'].str.replace(',', '').astype(float)

df['DistanceMiles'] = df['DistanceMiles'].str.replace(',', '').astype(float)
```

```python
df.head()
```

```python
df.isnull().sum()
```

there is no null values our data is clear already

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
lb =LabelEncoder()
```

```python
df['Destination'] =lb.fit_transform(df['Destination'])
df['DestCityName'] =lb.fit_transform(df['DestCityName'])
df['Origin'] =lb.fit_transform(df['Origin'])
df['OriginCityName'] =lb.fit_transform(df['OriginCityName'])
df['OriginWeather'] =lb.fit_transform(df['OriginWeather'])
df['DestCountry'] =lb.fit_transform(df['DestCountry'])
```

```python
df.head()
```

```python
x =df.drop(columns =['DestWeather'])
y =df['DestWeather']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test =train_test_split(x,y,test_size = 0.2,random_state =42)
```

```python
df.shape
```

```python
x_train.shape
```

```python
# Now we apply some regression to find the accuracy of the model 
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rf =RandomForestClassifier()
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred =rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test , y_pred)                  
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr =LogisticRegression()
```

```python
lr.fit(x_train ,y_train)
```

```python
y_pred =lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(LogisticRegression(),x,y,cv = 5,scoring ='accuracy'))
```

```python
# now we do hypertuning for the best parameters
# as we know that the random forest has 25 hyper parameters that's why we have  do fine tuning
```

```python
              # Gridsearch cv
```

```python
# numbers of trees in randoom forest
n_estimators = [20,60,100,120]

# numbers of features to consider at evry split(columns)
max_features =[0.2,0.6,1.0]

# maximum numbers of levels in tree
max_depth = [2,8,None]

# number of samples(rows)
max_samples = [0.5,0.75,1.0]

# 108 different random forest train
```

```python
param_grid ={'n_estimators' : n_estimators,
                 'max_features' : max_features,
                 'max_depth' : max_depth,
                 'max_samples' :max_samples
}

print(param_grid)
```

```python
rf =RandomForestClassifier()
```

```python
from sklearn.model_selection import GridSearchCV

rf_grid = GridSearchCV(estimator = rf,
                       param_grid = param_grid,
                       cv =5,
                       verbose = 2,
                       n_jobs = -1)
```

```python
rf_grid.fit(x_train, y_train)
```

```python
rf_grid.best_score_
```

```python
rf_grid.best_params_
```
